## Задача 3
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='reg');
plt.xlabel("Жилая площадь")
plt.ylabel("Стоимость недвижимости");
```

### Выводы
* График взаимного распределения жилой площади и стоимости не выявляет строгой линейной зависимости цены недвижимости от жилой площади хотя определенная взаимосвязь прослеживается применительно к объектам с жилой площадью до 6000

```python
state_price = df.groupby('condition')['price'].mean().reset_index()
plt.figure(figsize=(8, 6))
plt.title('Взаимосвязь состояния недвижимости и средней стоимости')
sns.barplot(x='condition', y='price', data=state_price)
plt.xlabel('состояние недвижимости')
plt.ylabel('средняя стоимость')
plt.show()
```

### Выводы
* прослеживается корелляция - чем лучше состояние недвижимсоти тем выше средняя стоимость, хотя и средняя стоимость домов в состоянии 1 чуть выше чем с состоянием 2, а средняя стоимость домов в состоянии 3 чуть выше чем с состоянием 4, что вероятно определяется влиянием и других параметров на стоимость недвижимости  

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['waterfront'], y=df['price'])
plt.xlabel('Вид на набережную')
plt.ylabel('Стоимость недвижимости')
plt.title('Распределение стоимости в зависимости от вида на набережную')
plt.xticks([0, 1], ["нет", "да"]);
plt.show();
```

### Выводы
* бокс, медиана и усы ящика недвижимости с видом на набережную находятся выше по оси стоимости, что явно указывает на то, что недвижимость с видом на набережную стоит дороже

```python
corr_matrix=df[['price','grade','floors','waterfront']].corr().round(2)
corr_matrix[np.abs(corr_matrix<0.3)]=0
sns.heatmap(corr_matrix,annot=True,linewidths=.5,cmap='coolwarm')
```

### Выводы
* тепловая карта четко демонстрирует, что стоимость недвижимости зависит прежде всего от grade - качества конструкции и дизайна, а не от количества этажей или вида на набережную

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='yr_built', y='price', data=df)
plt.xlabel('Год постройки')
plt.ylabel('Цена')
plt.title('Взаимосвязь между годом постройки и ценой')
plt.show()
```

### Выводы
* График рассеяния помогает визуализировать невыраженную общую тенденцию повышения стоимости домов построенных позже
